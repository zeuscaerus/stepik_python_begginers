–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ 2. –ü—Ä–æ–≤–µ—Ä—è—é—â–∞—è —Å–∏—Å—Ç–µ–º–∞ –ø–æ–ª—å–∑—É–µ—Ç—Å—è —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–º –≤—ã–≤–æ–¥–æ–º (stdout, –∫–æ–º–∞–Ω–¥–∞ print()).

–í –ø–æ–ø—É–ª—è—Ä–Ω–æ–º —Å–µ—Ä–∏–∞–ª–µ ¬´–û—Å—Ç–∞—Ç—å—Å—è –≤ –∂–∏–≤—ã—Ö¬ª –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–ª–∞—Å—å –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —á–∏—Å–µ–ª 4 8 15 16 23 42, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–µ—Å–ª–∞ –≥–µ—Ä–æ—è–º —É–¥–∞—á—É –∏ –ø–æ–º–æ–≥–ª–∞ —Å–æ—Ä–≤–∞—Ç—å –¥–∂–µ–∫–ø–æ—Ç –≤ –ª–æ—Ç–µ—Ä–µ–µ. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç –¥–∞–Ω–Ω—É—é –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —á–∏—Å–µ–ª —Å –æ–¥–Ω–∏–º –ø—Ä–æ–±–µ–ª–æ–º –º–µ–∂–¥—É –Ω–∏–º–∏.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –¢–µ–∫—Å—Ç '4 8 15 16 23 42' –Ω–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å. –í–æ—Å–ø–æ–ª—å–∑—É–π—Ç–µ—Å—å –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å—é –∫–æ–º–∞–Ω–¥—ã print() –≤—ã–≤–æ–¥–∏—Ç—å –Ω–µ—Å–∫–æ–ª—å–∫–æ –∞—Ä–≥—É–º–µ–Ω—Ç–æ–≤, —É–∫–∞–∑–∞–Ω–Ω—ã—Ö —á–µ—Ä–µ–∑ –∑–∞–ø—è—Ç—É—é.

```python
print('4', '8', '15', '16', '23', '42')
```

–°—á–∞—Å—Ç–ª–∏–≤–∞—è –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å 2

–ò–∑–º–µ–Ω–∏—Ç–µ –ø—Ä–µ–¥—ã–¥—É—â—É—é –ø—Ä–æ–≥—Ä–∞–º–º—É —Ç–∞–∫, —á—Ç–æ–±—ã –∫–∞–∂–¥–æ–µ —á–∏—Å–ª–æ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ 4 8 15 16 23 42 –ø–µ—á–∞—Ç–∞–ª–æ—Å—å –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –ö–∞–∂–¥–∞—è –ø–æ—Å–ª–µ–¥—É—é—â–∞—è –∫–æ–º–∞–Ω–¥–∞ print() –≤—ã–≤–æ–¥–∏—Ç —É–∫–∞–∑–∞–Ω–Ω—ã–π —Ç–µ–∫—Å—Ç, –Ω–∞—á–∏–Ω–∞—è —Å –Ω–æ–≤–æ–π —Å—Ç—Ä–æ–∫–∏.

```python
print('4')
print('8')
print('15')
print('16')
print('23')
print('42')
```

–ü—Ä–∏–≤–µ—Ç—Å—Ç–≤–∏–µ

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —Å—Ç—Ä–æ–∫–∞ —Ç–µ–∫—Å—Ç–∞ ‚Äì –∏–º—è —á–µ–ª–æ–≤–µ–∫–∞. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç –Ω–∞ —ç–∫—Ä–∞–Ω –ø—Ä–∏–≤–µ—Ç—Å—Ç–≤–∏–µ –≤ –≤–∏–¥–µ —Å–ª–æ–≤–∞ ¬´–ü—Ä–∏–≤–µ—Ç¬ª (–±–µ–∑ –∫–∞–≤—ã—á–µ–∫), –ø–æ—Å–ª–µ –∫–æ—Ç–æ—Ä–æ–≥–æ –¥–æ–ª–∂–Ω–∞ —Å—Ç–æ—è—Ç—å –∑–∞–ø—è—Ç–∞—è –∏ –ø—Ä–æ–±–µ–ª, –∞ –∑–∞—Ç–µ–º –≤–≤–µ–¥–µ–Ω–Ω–æ–µ –∏–º—è.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:
 –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –æ–¥–Ω–∞ —Å—Ç—Ä–æ–∫–∞ ‚Äî –∏–º—è —á–µ–ª–æ–≤–µ–∫–∞.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:
 –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç–µ–∫—Å—Ç –≤ —Å–æ–æ—Ç–≤–µ—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –î–ª—è —Å—á–∏—Ç—ã–≤–∞–Ω–∏—è —Ç–µ–∫—Å—Ç–∞ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ –∫–æ–º–∞–Ω–¥—É input(), –¥–ª—è –ø–µ—á–∞—Ç–∏ —Ç–µ–∫—Å—Ç–∞ –Ω–∞ —ç–∫—Ä–∞–Ω–µ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ –∫–æ–º–∞–Ω–¥—É print().

```python
print('–ü—Ä–∏–≤–µ—Ç, '+input())
```

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —Å—Ç—Ä–æ–∫–∞ —Ç–µ–∫—Å—Ç–∞ ‚Äì –Ω–∞–∑–≤–∞–Ω–∏–µ —Ñ—É—Ç–±–æ–ª—å–Ω–æ–π –∫–æ–º–∞–Ω–¥—ã. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø–æ–≤—Ç–æ—Ä—è–µ—Ç –µ–µ –Ω–∞ —ç–∫—Ä–∞–Ω–µ —Å–æ —Å–ª–æ–≤–∞–º–∏ ¬´ - —á–µ–º–ø–∏–æ–Ω!¬ª (–±–µ–∑ –∫–∞–≤—ã—á–µ–∫).

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:
 –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –Ω–∞–∑–≤–∞–Ω–∏–µ —Ñ—É—Ç–±–æ–ª—å–Ω–æ–π –∫–æ–º–∞–Ω–¥—ã.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –î–ª—è —Å—á–∏—Ç—ã–≤–∞–Ω–∏—è —Ç–µ–∫—Å—Ç–∞ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ –∫–æ–º–∞–Ω–¥—É input(), –¥–ª—è –ø–µ—á–∞—Ç–∏ —Ç–µ–∫—Å—Ç–∞ –Ω–∞ —ç–∫—Ä–∞–Ω–µ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ –∫–æ–º–∞–Ω–¥—É print().

```python
print(f'{input().capitalize()} - —á–µ–º–ø–∏–æ–Ω!')
```

–ü–æ–≤—Ç–æ—Ä—è–π –∑–∞ –º–Ω–æ–π

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å—á–∏—Ç—ã–≤–∞–µ—Ç —Ç—Ä–∏ —Å—Ç—Ä–æ–∫–∏ –ø–æ –æ—á–µ—Ä–µ–¥–∏, –∞ –∑–∞—Ç–µ–º –≤—ã–≤–æ–¥–∏—Ç –∏—Ö –≤ —Ç–æ–π –∂–µ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏, –∫–∞–∂–¥—É—é –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ—á–∫–µ.

 –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—é—Ç—Å—è —Ç—Ä–∏ —Å—Ç—Ä–æ–∫–∏, –∫–∞–∂–¥–∞—è –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

 –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –≤–≤–µ–¥–µ–Ω–Ω—ã–µ —Å—Ç—Ä–æ–∫–∏ –≤ —Ç–æ–π –∂–µ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏, –∫–∞–∂–¥—É—é –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –î–ª—è —Å—á–∏—Ç—ã–≤–∞–Ω–∏—è —Ç–µ–∫—Å—Ç–∞ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ –∫–æ–º–∞–Ω–¥—É input(), –¥–ª—è –ø–µ—á–∞—Ç–∏ —Ç–µ–∫—Å—Ç–∞ –Ω–∞ —ç–∫—Ä–∞–Ω–µ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ –∫–æ–º–∞–Ω–¥—É print().

```python
a, b, c=input(), input(), input()
print(a)
print(b)
print(c)
```

–ü–æ–≤—Ç–æ—Ä—è–π –∑–∞ –º–Ω–æ–π 2

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å—á–∏—Ç—ã–≤–∞–µ—Ç —Ç—Ä–∏ —Å—Ç—Ä–æ–∫–∏ –ø–æ –æ—á–µ—Ä–µ–¥–∏, –∞ –∑–∞—Ç–µ–º –≤—ã–≤–æ–¥–∏—Ç –∏—Ö –≤ –æ–±—Ä–∞—Ç–Ω–æ–π –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏, –∫–∞–∂–¥—É—é –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ—á–∫–µ.

 –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è —Ç—Ä–∏ —Å—Ç—Ä–æ–∫–∏, –∫–∞–∂–¥–∞—è –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ 3 –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è –≤–≤–µ–¥—ë–Ω–Ω—ã—Ö —Å—Ç—Ä–æ–∫ —Ç–µ–∫—Å—Ç–∞.

```python
a,b,c=input(),input(),input()
print(c,b,a, sep='\n')
```

2.3 –ü–∞—Ä–∞–º–µ—Ç—Ä—ã sep –∏ end

I like Python

–ü—Ä–∏–≤–µ—Ç—Å—Ç–≤–∏–µ

–ö—É–±

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –≤—ã—á–∏—Å–ª—è—é—â—É—é –æ–±—ä—ë–º –∫—É–±–∞ –∏ –ø–ª–æ—â–∞–¥—å –µ–≥–æ –ø–æ–ª–Ω–æ–π –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏, –ø–æ –≤–≤–µ–¥—ë–Ω–Ω–æ–º—É –∑–Ω–∞—á–µ–Ω–∏—é –¥–ª–∏–Ω—ã —Ä–µ–±—Ä–∞.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:
 –ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞–µ—Ç—Å—è –æ–¥–Ω–æ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ ‚Äì –¥–ª–∏–Ω–∞ —Ä–µ–±—Ä–∞.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:
 –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç–µ–∫—Å—Ç –∏ —á–∏—Å–ª–∞ –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏.

–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –û–±—ä—ë–º –∫—É–±–∞ –∏ –ø–ª–æ—â–∞–¥—å –ø–æ–ª–Ω–æ–π –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ –º–æ–∂–Ω–æ –≤—ã—á–∏—Å–ª–∏—Ç—å –ø–æ —Ñ–æ—Ä–º—É–ª–∞–º V = a^3, S = 6a^2.

```python
a=int(input())
print('–û–±—ä–µ–º =', a**3)
print('–ü–ª–æ—â–∞–¥—å –ø–æ–ª–Ω–æ–π –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ =', 6*a**2)
```

–ó–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –∑–Ω–∞—á–µ–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–∏ f(a,b)=3(a+b)^3+275b^2-127a-41 –ø–æ –≤–≤–µ–¥–µ–Ω—ã–º —Ü–µ–ª—ã–º –∑–µ–∞—á–µ–Ω–∏—è–º –∞ –∏ b.


–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –¥–≤–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ. –í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ ‚Äî –∑–Ω–∞—á–µ–Ω–∏–µ a, –≤–æ –≤—Ç–æ—Ä–æ–π —Å—Ç—Ä–æ–∫–µ ‚Äî –∑–Ω–∞—á–µ–Ω–∏–µ b.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –∑–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–æ –≤–≤–µ–¥—ë–Ω–Ω—ã–º —á–∏—Å–ª–∞–º a –∏ b.

```python
a, b = int(input()), int(input())
print(3*(a+b)**3+275*b**2-127*a-41)
```

–°–ª–µ–¥—É—é—â–µ–µ –∏ –ø—Ä–µ–¥—ã–¥—É—â–µ–µ

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å—á–∏—Ç—ã–≤–∞–µ—Ç —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ, –ø–æ—Å–ª–µ —á–µ–≥–æ –Ω–∞ —ç–∫—Ä–∞–Ω –≤—ã–≤–æ–¥–∏—Ç—Å—è —Å–ª–µ–¥—É—é—â–µ–µ –∏ –ø—Ä–µ–¥—ã–¥—É—â–µ–µ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ —Å –ø–æ—è—Å–Ω–∏—Ç–µ–ª—å–Ω—ã–º —Ç–µ–∫—Å—Ç–æ–º.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç–µ–∫—Å—Ç —Å–æ–≥–ª–∞—Å–Ω–æ —É—Å–ª–æ–≤–∏—é –∑–∞–¥–∞—á–∏.

```python
a=int(input())
print(f'–°–ª–µ–¥—É—é—â–µ–µ –∑–∞ —á–∏—Å–ª–æ–º {a} —á–∏—Å–ª–æ: {a+1}')
print(f'–î–ª—è —á–∏—Å–ª–∞ {a} –ø—Ä–µ–¥—ã–¥—É—â–µ–µ —á–∏—Å–ª–æ: {a-1}')
```

–°—Ç–æ–∏–º–æ—Å—Ç—å –ø–æ–∫—É–ø–∫–∏

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å—á–∏—Ç–∞–µ—Ç —Å—Ç–æ–∏–º–æ—Å—Ç—å —Ç—Ä–µ—Ö –∫–æ–º–ø—å—é—Ç–µ—Ä–æ–≤, —Å–æ—Å—Ç–æ—è—â–∏—Ö –∏–∑ –º–æ–Ω–∏—Ç–æ—Ä–∞, —Å–∏—Å—Ç–µ–º–Ω–æ–≥–æ –±–ª–æ–∫–∞, –∫–ª–∞–≤–∏–∞—Ç—É—Ä—ã –∏ –º—ã—à–∏.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è —á–µ—Ç—ã—Ä–µ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ. –í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ ‚Äî —Å—Ç–æ–∏–º–æ—Å—Ç—å –º–æ–Ω–∏—Ç–æ—Ä–∞, –≤–æ –≤—Ç–æ—Ä–æ–π —Å—Ç—Ä–æ–∫–µ ‚Äî —Å—Ç–æ–∏–º–æ—Å—Ç—å —Å–∏—Å—Ç–µ–º–Ω–æ–≥–æ –±–ª–æ–∫–∞, –≤ —Ç—Ä–µ—Ç—å–µ–π —Å—Ç—Ä–æ–∫–µ ‚Äî —Å—Ç–æ–∏–º–æ—Å—Ç—å –∫–ª–∞–≤–∏–∞—Ç—É—Ä—ã –∏ –≤ —á–µ—Ç–≤–µ—Ä—Ç–æ–π —Å—Ç—Ä–æ–∫–µ ‚Äî —Å—Ç–æ–∏–º–æ—Å—Ç—å –º—ã—à–∏.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ–¥–Ω–æ —á–∏—Å–ª–æ ‚Äì —Å—Ç–æ–∏–º–æ—Å—Ç—å –ø–æ–∫—É–ø–∫–∏ (—Ç—Ä–µ—Ö –∫–æ–º–ø—å—é—Ç–µ—Ä–æ–≤).

–ê—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∏–µ –æ–ø–µ—Ä–∞—Ü–∏–∏

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –≤ –∫–æ—Ç–æ—Ä–æ–π –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è —Å—É–º–º–∞, —Ä–∞–∑–Ω–æ—Å—Ç—å –∏ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –¥–≤—É—Ö —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª, –≤–≤–µ–¥–µ–Ω–Ω—ã—Ö —Å –∫–ª–∞–≤–∏–∞—Ç—É—Ä—ã.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –¥–≤–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Å—É–º–º—É, —Ä–∞–∑–Ω–æ—Å—Ç—å –∏ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –≤–≤–µ–¥—ë–Ω–Ω—ã—Ö —á–∏—Å–µ–ª, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

```python
a, b=int(input()), int(input())
print(f'{a} + {b} = {a+b}')
print(f'{a} - {b} = {a-b}')
print(f'{a} * {b} = {a*b}')
```

–ê—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∞—è –ø—Ä–æ–≥—Ä–µ—Å—Å–∏—è

–ê—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–π –ø—Ä–æ–≥—Ä–µ—Å—Å–∏–µ–π –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —á–∏—Å–µ–ª a1,a2,...,an, –∫–∞–∂–¥–æ–µ –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö, –Ω–∞—á–∏–Ω–∞—è —Å a2,–ø–æ–ª—É—á–∞–µ—Ç—Å—è –∏–∑ –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ –ø—Ä–∏–±–∞–≤–ª–µ–Ω–∏–µ–º –∫ –Ω–µ–º—É –æ–¥–Ω–æ–≥–æ –∏ —Ç–æ–≥–æ –∂–µ –ø–æ—Å—Ç–æ—è–Ω–Ω–æ–≥–æ —á–∏—Å–ª–∞ d (—Ä–∞–∑–Ω–æ—Å—Ç—å –ø—Ä–æ–≥—Ä–µ—Å—Å–∏–∏), —Ç–æ –µ—Å—Ç—å: an=an-1+d

–ï—Å–ª–∏ –∏–∑–≤–µ—Å—Ç–µ–Ω –ø–µ—Ä–≤—ã–π —á–ª–µ–Ω –ø—Ä–æ–≥—Ä–µ—Å—Å–∏–∏ –∏ –µ—ë —Ä–∞–∑–Ω–æ—Å—Ç—å, —Ç–æ n-—ã–π —á–ª–µ–Ω –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–π –ø—Ä–æ–≥—Ä–µ—Å—Å–∏–∏ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –ø–æ —Ñ–æ—Ä–º—É–ª–µ: an=a1+d(n-1)

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è —Ç—Ä–∏ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞: a1, d –∏ n, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ n-—ã–π —á–ª–µ–Ω –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–π –ø—Ä–æ–≥—Ä–µ—Å—Å–∏–∏.

```python
a = int(input())
d = int(input())
n = int(input())
print(a+d*(n-1))
```

–†–∞–∑–¥–µ–ª—è–π –∏ –≤–ª–∞—Å—Ç–≤—É–π

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Å—á–∏—Ç—ã–≤–∞–µ—Ç —Ü–µ–ª–æ–µ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ x –∏ –≤—ã–≤–æ–¥–∏—Ç –Ω–∞ —ç–∫—Ä–∞–Ω –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —á–∏—Å–µ–ª x, 2x, 3x, 4x –∏ 5x, —Ä–∞–∑–¥–µ–ª—ë–Ω–Ω—ã—Ö —Ç—Ä–µ–º—è —á–µ—Ä—Ç–æ—á–∫–∞–º–∏.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è —Ü–µ–ª–æ–µ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ —á–∏—Å–ª–æ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç–µ–∫—Å—Ç —Å–æ–≥–ª–∞—Å–Ω–æ —É—Å–ª–æ–≤–∏—é –∑–∞–¥–∞—á–∏.

```python
a = int(input())
print(a,2*a,3*a,4*a,5*a, sep='---')
```

–ì–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–∞—è –ø—Ä–æ–≥—Ä–µ—Å—Å–∏—è

–ì–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–π –ø—Ä–æ–≥—Ä–µ—Å—Å–∏–µ–π –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —á–∏—Å–µ–ª b1, b2,‚Ä¶,bn, –∫–∞–∂–¥–æ–µ –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö, –Ω–∞—á–∏–Ω–∞—è —Å b2, –ø–æ–ª—É—á–∞–µ—Ç—Å—è –∏–∑ –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ —É–º–Ω–æ–∂–µ–Ω–∏–µ–º –Ω–∞ –æ–¥–Ω–æ –∏ —Ç–æ –∂–µ –ø–æ—Å—Ç–æ—è–Ω–Ω–æ–µ —á–∏—Å–ª–æ q (–∑–Ω–∞–º–µ–Ω–∞—Ç–µ–ª—å –ø—Ä–æ–≥—Ä–µ—Å—Å–∏–∏), —Ç–æ –µ—Å—Ç—å bn=b(n‚àí1)*q

–ï—Å–ª–∏ –∏–∑–≤–µ—Å—Ç–µ–Ω –ø–µ—Ä–≤—ã–π —á–ª–µ–Ω –ø—Ä–æ–≥—Ä–µ—Å—Å–∏–∏ –∏ –µ—ë –∑–Ω–∞–º–µ–Ω–∞—Ç–µ–ª—å, —Ç–æ n-—ã–π —á–ª–µ–Ω –≥–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–π –ø—Ä–æ–≥—Ä–µ—Å—Å–∏–∏ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –ø–æ —Ñ–æ—Ä–º—É–ª–µ bn=b1*q^n‚àí1

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è —Ç—Ä–∏ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞: b1, q –∏ n, –∫–∞–∂–¥–æ–µ –Ω–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–π —Å—Ç—Ä–æ–∫–µ.

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ n-—ã–π —á–ª–µ–Ω –≥–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–π –ø—Ä–æ–≥—Ä–µ—Å—Å–∏–∏.

```python
b,q,n = int(input()), int(input()), int(input())
print(b*q**(n-1))
```

–°–∞–º–∞ –Ω–µ–æ—Ç–≤—Ä–∞—Ç–∏–º–æ—Å—Ç—å üå∂Ô∏è

–ë–µ–∑—É–º–Ω—ã–π —Ç–∏—Ç–∞–Ω –¢–∞–Ω–æ—Å —Å–æ–±—Ä–∞–ª –≤—Å–µ 6 –∫–∞–º–Ω–µ–π –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ—Å—Ç–∏ –∏ –Ω–∞–º–µ—Ä–µ–≤–∞–µ—Ç—Å—è —É–Ω–∏—á—Ç–æ–∂–∏—Ç—å –ø–æ–ª–æ–≤–∏–Ω—É –Ω–∞—Å–µ–ª–µ–Ω–∏—è –í—Å–µ–ª–µ–Ω–Ω–æ–π –ø–æ —â–µ–ª—á–∫—É –ø–∞–ª—å—Ü–µ–≤. –ü—Ä–∏ —ç—Ç–æ–º –µ—Å–ª–∏ –Ω–∞—Å–µ–ª–µ–Ω–∏–µ –í—Å–µ–ª–µ–Ω–Ω–æ–π —è–≤–ª—è–µ—Ç—Å—è –Ω–µ—á–µ—Ç–Ω—ã–º —á–∏—Å–ª–æ–º, —Ç–æ —Ç–∏—Ç–∞–Ω –ø—Ä–æ—è–≤–∏—Ç –º–∏–ª–æ—Å–µ—Ä–¥–∏–µ –∏ –æ–∫—Ä—É–≥–ª–∏—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–∂–∏–≤—à–∏—Ö –≤ –±–æ–ª—å—à—É—é —Å—Ç–æ—Ä–æ–Ω—É. –ü–æ–º–æ–≥–∏—Ç–µ –ú—Å—Ç–∏—Ç–µ–ª—è–º –ø–æ–¥—Å—á–∏—Ç–∞—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–∂–∏–≤—à–∏—Ö.

–ù–∞ –≤—Ö–æ–¥ –¥–∞–µ—Ç—Å—è —á–∏—Å–ª–æ —Ü–µ–ª–æ–µ n ‚Äì –Ω–∞—Å–µ–ª–µ–Ω–∏–µ –í—Å–µ–ª–µ–Ω–Ω–æ–π.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ–¥–Ω–æ —á–∏—Å–ª–æ ‚Äì –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–∂–∏–≤—à–∏—Ö.

–ù–æ–º–µ—Ä –∫—É–ø–µ üå∂Ô∏è

–í –∫—É–ø–µ–π–Ω–æ–º –≤–∞–≥–æ–Ω–µ –∏–º–µ–µ—Ç—Å—è 9 –∫—É–ø–µ —Å —á–µ—Ç—ã—Ä—å–º—è –º–µ—Å—Ç–∞–º–∏ –¥–ª—è –ø–∞—Å—Å–∞–∂–∏—Ä–æ–≤ –≤ –∫–∞–∂–¥–æ–º. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç –Ω–æ–º–µ—Ä –∫—É–ø–µ, –≤ –∫–æ—Ç–æ—Ä–æ–º –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –º–µ—Å—Ç–æ —Å –∑–∞–¥–∞–Ω–Ω—ã–º –Ω–æ–º–µ—Ä–æ–º (–Ω—É–º–µ—Ä–∞—Ü–∏—è –º–µ—Å—Ç —Å–∫–≤–æ–∑–Ω–∞—è, –Ω–∞—á–∏–Ω–∞–µ—Ç—Å—è —Å 1).

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ ‚Äì –º–µ—Å—Ç–æ —Å –∑–∞–¥–∞–Ω–Ω—ã–º –Ω–æ–º–µ—Ä–æ–º –≤ –≤–∞–≥–æ–Ω–µ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –æ–¥–Ω–æ —á–∏—Å–ª–æ ‚Äì –Ω–æ–º–µ—Ä –∫—É–ø–µ, –≤ –∫–æ—Ç–æ—Ä–æ–º –Ω–∞—Ö–æ–¥–∏—Ç—Å—è —É–∫–∞–∑–∞–Ω–æ–µ –º–µ—Å—Ç–æ.

–ü–µ—Ä–µ—Å—á–µ—Ç –≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É –¥–ª—è –ø–µ—Ä–µ—Å—á—ë—Ç–∞ –≤–µ–ª–∏—á–∏–Ω—ã –≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞, –∑–∞–¥–∞–Ω–Ω–æ–≥–æ –≤ –º–∏–Ω—É—Ç–∞—Ö, –≤ –≤–µ–ª–∏—á–∏–Ω—É, –≤—ã—Ä–∞–∂–µ–Ω–Ω—É—é –≤ —á–∞—Å–∞—Ö –∏ –º–∏–Ω—É—Ç–∞—Ö.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ ‚Äì –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –º–∏–Ω—É—Ç.

–¢—Ä–µ—Ö–∑–Ω–∞—á–Ω–æ–µ —á–∏—Å–ª–æ

–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –≤ –∫–æ—Ç–æ—Ä–æ–π —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ—Ç—Å—è —Å—É–º–º–∞ –∏ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ —Ü–∏—Ñ—Ä –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ —Ç—Ä—ë—Ö–∑–Ω–∞—á–Ω–æ–≥–æ —á–∏—Å–ª–∞.

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ —Ç—Ä—ë—Ö–∑–Ω–∞—á–Ω–æ–µ —á–∏—Å–ª–æ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ –¥–≤–∞ —á–∏—Å–ª–∞ —Å –ø–æ—è—Å–Ω—è—é—â–∏–º —Ç–µ–∫—Å—Ç–æ–º: —Å—É–º–º–∞ —Ü–∏—Ñ—Ä –∏ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ —Ü–∏—Ñ—Ä.

```python
num = int(input())
a = num % 10
b = (num % 100) // 10
c = num // 100
print(f'–°—É–º–º–∞ —Ü–∏—Ñ—Ä = {a+b+c}')
print(f'–ü—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ —Ü–∏—Ñ—Ä = {a*b*c}')
```

–ü–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∞ —Ü–∏—Ñ—Ä
–î–∞–Ω–æ —Ç—Ä–µ—Ö–∑–Ω–∞—á–Ω–æ–µ —á–∏—Å–ª–æ abc, –≤ –∫–æ—Ç–æ—Ä–æ–º –≤—Å–µ —Ü–∏—Ñ—Ä—ã —Ä–∞–∑–ª–∏—á–Ω—ã. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–æ–¥–∏—Ç —à–µ—Å—Ç—å —á–∏—Å–µ–ª, –æ–±—Ä–∞–∑–æ–≤–∞–Ω–Ω—ã—Ö –ø—Ä–∏ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–µ —Ü–∏—Ñ—Ä –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —á–∏—Å–ª–∞.

–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ —Ç—Ä—ë—Ö–∑–Ω–∞—á–Ω–æ–µ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ, –≤—Å–µ —Ü–∏—Ñ—Ä—ã –∫–æ—Ç–æ—Ä–æ–≥–æ —Ä–∞–∑–ª–∏—á–Ω—ã.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —à–µ—Å—Ç—å —á–∏—Å–µ–ª, –æ–±—Ä–∞–∑–æ–≤–∞–Ω–Ω—ã—Ö –ø—Ä–∏ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–µ —Ü–∏—Ñ—Ä –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —á–∏—Å–ª–∞ –≤ —Å–ª–µ–¥—É—é—â–µ–º –ø–æ—Ä—è–¥–∫–µ: abc,acb,bac,bca,cab,cba.

–ß–µ—Ç—ã—Ä—ë—Ö–∑–Ω–∞—á–Ω–æ–µ —á–∏—Å–ª–æ

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: 
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ —á–µ—Ç—ã—Ä—ë—Ö–∑–Ω–∞—á–Ω–æ–µ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ.
